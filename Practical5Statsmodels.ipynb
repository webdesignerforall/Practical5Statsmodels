```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('california_housing_train.csv')
```

```python
df
```

```python
pop=df['population']
bedrooms=df['total_bedrooms']
```

```python
pop.shape
```

```python
bedrooms.shape
```

```python
plt.scatter(pop,bedrooms)
plt.xlabel('Population')
plt.ylabel('Total bedrooms')
plt.show()
```

```python
import statsmodels.api as sm
```

```python
pop_np=pop.to_numpy()
bedrooms_np=bedrooms.to_numpy()
```

```python
model=sm.OLS(bedrooms_np,pop_np).fit()
bedrooms_predictions=model.predict(pop_np)
plt.scatter(pop,bedrooms)
plt.xlabel('Population')
plt.ylabel('Total bedrooms')
plt.plot(pop,bedrooms_predictions)
```
